```python
import pandas as pd
import numpy as np

dataframe_alunos =  pd.read_csv('data/DM_ALUNO.csv', sep='|', encoding='latin1',
                                     usecols=['CO_IES', 'DS_CATEGORIA_ADMINISTRATIVA', 'NO_CURSO', 'DS_TURNO_ALUNO',
                                              'DS_MODALIDADE_ENSINO', 'NO_OCDE', 'NO_OCDE_AREA_GERAL', 'DS_SEXO_ALUNO'])
                                     
dataframe_iess = pd.read_csv('data/DM_IES.csv', sep='|', encoding='latin1',
                            usecols=['CO_IES', 'SGL_IES', 'NO_IES', 'NO_MUNICIPIO_IES', 'SGL_UF_IES', 'NO_REGIAO_IES'])

dataframe_docentes = pd.read_csv('data/DM_DOCENTE.csv', sep='|', encoding='latin1',
                                usecols=['CO_IES', 'DS_SEXO_DOCENTE'])
dataframe_alunos.head(10)
```

```python
#dataframe_bens['cd_tipo_bem'] = dataframe_bens.cd_tipo_bem.astype('category')
dataframe_docentes['DS_SEXO_DOCENTE'] = dataframe_docentes.DS_SEXO_DOCENTE.astype('category')

dataframe_iess['SGL_IES'] = dataframe_iess.SGL_IES.astype('category')
dataframe_iess['SGL_UF_IES'] = dataframe_iess.SGL_UF_IES.astype('category')
dataframe_iess['NO_REGIAO_IES'] = dataframe_iess.NO_REGIAO_IES.astype('category')

dataframe_alunos['DS_CATEGORIA_ADMINISTRATIVA'] = dataframe_alunos.DS_CATEGORIA_ADMINISTRATIVA.astype('category')
dataframe_alunos['DS_TURNO_ALUNO'] = dataframe_alunos.DS_TURNO_ALUNO.astype('category')
dataframe_alunos['DS_MODALIDADE_ENSINO'] = dataframe_alunos.DS_MODALIDADE_ENSINO.astype('category')
dataframe_alunos['DS_SEXO_ALUNO'] = dataframe_alunos.DS_SEXO_ALUNO.astype('category')
dataframe_alunos['NO_OCDE_AREA_GERAL'] = dataframe_alunos.NO_OCDE_AREA_GERAL.astype('category')

dataframe_iess.head(10)
```

```python
df_docentes_fem = dataframe_docentes.loc[dataframe_docentes['DS_SEXO_DOCENTE'] == 'Feminino']
df_docentes_masc = dataframe_docentes.loc[dataframe_docentes['DS_SEXO_DOCENTE'] == 'Masculino']

df_docentes_fem = df_docentes_fem.groupby('CO_IES', as_index=False).count()
df_docentes_masc = df_docentes_masc.groupby('CO_IES', as_index=False).count()
#tb = tb.rename(columns={'cd_tipo_bem': 'quantidade_bens_declarados'})
df_docentes_fem = df_docentes_fem.rename(columns={'DS_SEXO_DOCENTE': 'DOCENTES_GEN_FEM'})
df_docentes_masc = df_docentes_masc.rename(columns={'DS_SEXO_DOCENTE': 'DOCENTES_GEN_MASC'})
df_docentes_fem.fillna(0, inplace=True)
df_docentes_masc.fillna(0, inplace=True)
#result = pd.merge(dataframe_candidatos, tb, how='outer', on=['sq_candidato'])
dataframe_iess = pd.merge(dataframe_iess, df_docentes_fem, how='outer', on=['CO_IES'])
dataframe_iess = pd.merge(dataframe_iess, df_docentes_masc, how='outer', on=['CO_IES'])
dataframe_iess['DOCENTES_GEN_FEM'] = dataframe_iess['DOCENTES_GEN_FEM'].fillna(0) 
dataframe_iess['DOCENTES_GEN_MASC'] = dataframe_iess['DOCENTES_GEN_MASC'].fillna(0) 
#FILLNA EM DOECNTES
dataframe_iess
```

```python
#teste = dataframe_iess.loc[pd.isna(dataframe_iess['DOCENTES_GEN_MASC']) | pd.isna(dataframe_iess['DOCENTES_GEN_FEM'])]
#teste = dataframe_docentes.loc[dataframe_docentes['CO_IES']==1706]
#print(dataframe_docentes.dtypes)
df = dataframe_iess[['NO_IES', 'SGL_IES']]
#df['Cat1'].fillna(df['Cat2'])
#df.Cat1 = np.where(df.Cat1.isnull(), df.Cat2, df.Cat1)
df.SGL_IES = np.where(df.SGL_IES.isnull(), df.NO_IES, df.SGL_IES)
dataframe_iess.SGL_IES = df.SGL_IES
dataframe_iess['DOCENTES_TOTAL'] = dataframe_iess['DOCENTES_GEN_FEM'] + dataframe_iess['DOCENTES_GEN_MASC'] 
#float("{0:.2f}".format(13.949999999999999))
dataframe_iess['PORCENTAGEM_DOCENTES_FEM'] = (dataframe_iess['DOCENTES_GEN_FEM'] / dataframe_iess['DOCENTES_TOTAL'])*100
#np.round(p_table, decimals=2)
dataframe_iess.PORCENTAGEM_DOCENTES_FEM = np.round(dataframe_iess.PORCENTAGEM_DOCENTES_FEM, decimals=2)
dataframe_iess
```

```python
dataframe_alunos_fem = dataframe_alunos.loc[dataframe_alunos['DS_SEXO_ALUNO'] == 'Feminino']
dataframe_alunos_masc = dataframe_alunos.loc[dataframe_alunos['DS_SEXO_ALUNO'] == 'Masculino']


dataframe_alunos_fem = dataframe_alunos_fem.groupby('CO_IES', as_index=False).count()
dataframe_alunos_masc = dataframe_alunos_masc.groupby('CO_IES', as_index=False).count()

dataframe_alunos_fem = dataframe_alunos_fem[['CO_IES', 'NO_OCDE']]
dataframe_alunos_masc = dataframe_alunos_masc[['CO_IES', 'NO_OCDE']]
#tb = tb.rename(columns={'cd_tipo_bem': 'quantidade_bens_declarados'})

dataframe_alunos_fem = dataframe_alunos_fem.rename(columns={'NO_OCDE': 'ALUNOS_GEN_FEM'})
dataframe_alunos_masc = dataframe_alunos_masc.rename(columns={'NO_OCDE': 'ALUNOS_GEN_MASC'})

dataframe_alunos_fem.fillna(0, inplace=True)
dataframe_alunos_masc.fillna(0, inplace=True)

dataframe_iess = pd.merge(dataframe_iess, dataframe_alunos_fem, how='outer', on=['CO_IES'])
dataframe_iess = pd.merge(dataframe_iess, dataframe_alunos_masc, how='outer', on=['CO_IES'])
dataframe_iess['ALUNOS_GEN_FEM'] = dataframe_iess['ALUNOS_GEN_FEM'].fillna(0) 
dataframe_iess['ALUNOS_GEN_MASC'] = dataframe_iess['ALUNOS_GEN_MASC'].fillna(0) 

dataframe_alunos_masc.loc[pd.isna(dataframe_alunos_masc.ALUNOS_GEN_MASC)]
```

```python
dataframe_iess['ALUNOS_TOTAL'] = dataframe_iess['ALUNOS_GEN_FEM'] + dataframe_iess['ALUNOS_GEN_MASC'] 
dataframe_iess['PORCENTAGEM_ALUNOS_FEM'] = (dataframe_iess['ALUNOS_GEN_FEM'] / dataframe_iess['ALUNOS_TOTAL'])*100
dataframe_iess.PORCENTAGEM_ALUNOS_FEM = np.round(dataframe_iess.PORCENTAGEM_ALUNOS_FEM, decimals=2)
dataframe_iess 
```

```python
exatas = ['Ciências, matemática e computação', 'Engenharia, produção e construção']
biologicas = ['Agricultura e veterinária', 'Saúde e bem estar social']
humanas = ['Ciências sociais, negócios e direito', 'Educação', 'Humanidades e artes', 'Serviços']
```

```python
#teste = dataframe_alunos.groupby(['CO_IES', 'NO_OCDE', 'DS_SEXO_ALUNO'], as_index=True).count()
dataframe_fem = dataframe_alunos.loc[dataframe_alunos['DS_SEXO_ALUNO'] == 'Feminino']
dataframe_mas = dataframe_alunos.loc[dataframe_alunos['DS_SEXO_ALUNO'] == 'Masculino']
#area 2: Ciências, matemática e computação , Engenharia, produção e construção
#area 3: Agricultura e veterinária , Saúde e bem estar social
#area 1: Ciências sociais, negócios e direito , Educação , Humanidades e artes , Serviços
#dataframe_imoveis = dataframe_bens[dataframe_bens.ds_tipo_bem.isin(imovel)]
dataframe_fem_exatas = dataframe_fem[dataframe_fem.NO_OCDE_AREA_GERAL.isin(exatas)]
dataframe_masc_exatas = dataframe_mas[dataframe_mas.NO_OCDE_AREA_GERAL.isin(exatas)]

dataframe_fem_biologicas = dataframe_fem[dataframe_fem.NO_OCDE_AREA_GERAL.isin(biologicas)]
dataframe_masc_biologicas = dataframe_mas[dataframe_mas.NO_OCDE_AREA_GERAL.isin(biologicas)]

dataframe_fem_humanas = dataframe_fem[dataframe_fem.NO_OCDE_AREA_GERAL.isin(humanas)]
dataframe_masc_humanas = dataframe_mas[dataframe_mas.NO_OCDE_AREA_GERAL.isin(humanas)]

dataframe_alunos.groupby('NO_OCDE_AREA_GERAL').count()
```

```python
dataframe_fem_exatas = dataframe_fem_exatas.groupby('CO_IES', as_index=False).count()
dataframe_masc_exatas = dataframe_masc_exatas.groupby('CO_IES', as_index=False).count()

dataframe_fem_biologicas = dataframe_masc_biologicas.groupby('CO_IES', as_index=False).count()
dataframe_masc_biologicas = dataframe_masc_biologicas.groupby('CO_IES', as_index=False).count()

dataframe_fem_humanas = dataframe_fem_humanas.groupby('CO_IES', as_index=False).count()
dataframe_masc_humanas = dataframe_masc_humanas.groupby('CO_IES', as_index=False).count()


dataframe_fem_exatas.fillna(0, inplace=True) 
dataframe_masc_exatas.fillna(0, inplace=True) 

dataframe_fem_biologicas.fillna(0, inplace=True) 
dataframe_masc_biologicas.fillna(0, inplace=True) 

dataframe_fem_humanas.fillna(0, inplace=True) 
dataframe_masc_humanas.fillna(0, inplace=True) 


dataframe_fem_exatas = dataframe_fem_exatas[['CO_IES', 'NO_OCDE']]
dataframe_masc_exatas = dataframe_masc_exatas[['CO_IES', 'NO_OCDE']]

dataframe_fem_biologicas = dataframe_fem_biologicas[['CO_IES', 'NO_OCDE']]
dataframe_masc_biologicas = dataframe_masc_biologicas[['CO_IES', 'NO_OCDE']]

dataframe_fem_humanas = dataframe_fem_humanas[['CO_IES', 'NO_OCDE']]
dataframe_masc_humanas = dataframe_masc_humanas[['CO_IES', 'NO_OCDE']]

dataframe_fem_exatas = dataframe_fem_exatas.rename(columns={'NO_OCDE': 'ALUNOS_FEM_EXATAS'})
dataframe_masc_exatas = dataframe_masc_exatas.rename(columns={'NO_OCDE': 'ALUNOS_MASC_EXATAS'})

dataframe_fem_biologicas = dataframe_fem_biologicas.rename(columns={'NO_OCDE': 'ALUNOS_FEM_BIOLOGICAS'})
dataframe_masc_biologicas = dataframe_masc_biologicas.rename(columns={'NO_OCDE': 'ALUNOS_MASC_BIOLOGICAS'})

dataframe_fem_humanas = dataframe_fem_humanas.rename(columns={'NO_OCDE': 'ALUNOS_FEM_HUMANAS'})
dataframe_masc_humanas = dataframe_masc_humanas.rename(columns={'NO_OCDE': 'ALUNOS_MASC_HUMANAS'})

```

```python
dataframe_iess = pd.merge(dataframe_iess, dataframe_fem_exatas, how='outer', on=['CO_IES'])
dataframe_iess = pd.merge(dataframe_iess, dataframe_masc_exatas, how='outer', on=['CO_IES'])
dataframe_iess = pd.merge(dataframe_iess, dataframe_fem_biologicas, how='outer', on=['CO_IES'])
dataframe_iess = pd.merge(dataframe_iess, dataframe_masc_biologicas, how='outer', on=['CO_IES'])
dataframe_iess = pd.merge(dataframe_iess, dataframe_fem_humanas, how='outer', on=['CO_IES'])
dataframe_iess = pd.merge(dataframe_iess, dataframe_masc_humanas, how='outer', on=['CO_IES'])
dataframe_iess['ALUNOS_FEM_EXATAS'] = dataframe_iess['ALUNOS_FEM_EXATAS'].fillna(0)
dataframe_iess['ALUNOS_MASC_EXATAS'] = dataframe_iess['ALUNOS_MASC_EXATAS'].fillna(0)
dataframe_iess
```

```python
dataframe_iess['PORCENTAGEM_FEM_EXATAS'] = np.round((dataframe_iess['ALUNOS_FEM_EXATAS']/(dataframe_iess['ALUNOS_FEM_EXATAS']+dataframe_iess['ALUNOS_MASC_EXATAS']))*100, decimals=2)
dataframe_iess
```
